```python
import pandas as pd
from pandas_profiling import ProfileReport

import altair as alt
alt.renderers.enable('default')
```

```python
loans_df = pd.read_excel('/data/taller2.xlsx')
```

```python
loans_df.shape
```

```python
loans_df.dtypes
```

```python
loans_df.head()
```

Filtro por los parsonas que han sido identificados con riesgo de incumplimiento de la obligaci√≥n

```python
loans_df_risk_flag = loans_df[loans_df['risk_flag'] == 1]
```

Por restrincion de la libreria no se puede trabajar con mas de 5000 mil datos por lo tanto se crea este filtro la analizar una cantidad en especifico 


```python
loans_df_5000 = loans_df.iloc[:5000,:]
```

```python
alt.Chart(loans_df_5000).mark_boxplot().encode(
        alt.X('experience'), 
        alt.Y('income'), 
    ).properties(
        width = 600,
        height = 500
    ).interactive()
   
```

```python
alt.Chart(loans_df_5000).mark_point().encode(
        alt.X('risk_flag'), 
        alt.Y('income'), 
    ).properties(
        width = 600,
        height = 500
    ).interactive()
```

```python
alt.Chart(loans_df_5000).mark_point().encode(    
    alt.X('count():Q'),                    
    alt.Y('age:N', sort = '-x'), 
    alt.Color('age:N'),  
    tooltip = [alt.Tooltip('id'),                
               alt.Tooltip('age'),
               alt.Tooltip('income'),
               alt.Tooltip('experience'),
               alt.Tooltip('marital_status'),
               alt.Tooltip('house_ownership'),
               alt.Tooltip('car_ownership'),
               alt.Tooltip('city'),
               alt.Tooltip('state'),
			         alt.Tooltip('current_job_years'),
			         alt.Tooltip('current_house_years'),
               alt.Tooltip('risk_flag')
              ]
).properties(
    width = 1000,
    height = 50
).interactive()
```

```python
alt.Chart(loans_df_risk_flag).mark_bar().encode(    # Making a bar chart
    alt.X('count():Q'),                    # X-axis is an aggregated value for categorical variable in Y
    alt.Y('state:N', sort = '-x'),   # Y-axis is sorted based on aggregated value
    alt.Color('state:N'),
    tooltip = [alt.Tooltip('count():Q')]
).properties(
    width = 600,
    height = 450
)
```

Usaremos boxplot para mirar en un rango determinado la edad de quines presentan mas reportes

```python
alt.Chart(loans_df_risk_flag).mark_boxplot().encode(
    alt.X('age')
).properties(
    width = 1250,
    height = 100
)
```

Edad media de las personas reportadas

```python
loans_grouped_df = loans_df_risk_flag.groupby('risk_flag')\
    .agg({'age': 'mean'})\
    .reset_index()\
    .melt(id_vars = ['risk_flag'], var_name = 'Metric', value_name = 'Value' )\
    .sort_values(by = ['risk_flag', 'Metric'])
```

Detalle de la edad media

```python
loans_grouped_df.tail()
```
